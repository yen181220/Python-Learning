# Pyplot API vs Object-Oriented API


- pyplot : quickly, easy visualization
- oo api: advanced


```python
#pyplot eg:
x = np.linspace(1,10,1000)
plt.plot(x,np.sin(x),color = "White",linestyle = "dashed",label = "sin(x)")
plt.plot(x,np.cos(x),label = 'cos(x)')
plt.xlabel("Bien x")
plt.ylabel('Six(x)')
#plt.xlim([0,4])
#plt.ylim([-0.75,0.75])
plt.legend();
```

```python
#orientedobjectapieg
#step1 : Prepare the data
x=[1,2,3,4]
y=[11,22,33,44]
#step2: Set up plot
fig,ax = plt.subplots(figsize = (5,5)) #figsize of plot = width*height
#step3: Plot the data
ax.plot(x,y)
ax.set(title = "A simple plot",
       xlabel='x-axis',
       ylabel='y-axis');
```

# Line


```python
#create an array
x = np.linspace(0,10,100)
x
```

```python
#OO API plot a line chart
fig, ax = plt.subplots()
ax.plot(x,x**3);
```

Scatter plot: Biểu đồ phân tán


```python
#Pyplot API
plt.scatter(x,np.exp(x)); #x;y=e^x
```

```python
#OO API
#new data
rng = np.random.RandomState(0)
x = rng.random(100)
y = rng.randn(100)
fig,ax = plt.subplots()
ax.scatter(x,y,edgecolors='white');
#thay đổi size của chấm
sizes = 1000*rng.rand(100)
ax.scatter(x,y,s=sizes)
#thay đổi màu của chấm
colors = rng.rand(100)
ax.scatter(x,y,s=sizes,c=colors);
#thay đổi thêm nhiều màu sắc
ax.scatter(x,y,s = sizes, c = colors, cmap = 'viridis')
#xem thang màu
img1 = ax.scatter(x,y,s=sizes,c=colors,cmap = 'viridis')
fig.colorbar(img1);
```

# Bar chart


- **Vertical**
- **Horizontal**

```python
#Vertical (Column chart)
#prepare data
soft_drinks_prices = {"Coke":10,
                      "Pepsi":12,
                      "Sprite":8}
fig,ax = plt.subplots()
ax.bar(soft_drinks_prices.keys(),soft_drinks_prices.values())
ax.set(title = "Soft Drinks Prices",
       xlabel = "Drinks",
       ylabel = "Prices");
```

```python
#Horizontal = bar
fig,ax = plt.subplots()
ax.barh(list(soft_drinks_prices.keys()),list(soft_drinks_prices.values()))
```

# Histogram

```python
#prepare data
np.random.seed(42)
student_height = np.random.normal(170,10,250)
student_height
```

```python
#pyplot api
plt.hist(student_height);
#chia nhỏ trục y
plt.hist(student_height,bins=30);
```

# Subplots

Multiple plot on one figure

```python
#Option1
plt.style.use('classic')
fig,((ax1,ax2),(ax3,ax4))= plt.subplots(nrows = 2,
                            ncols = 2,
                            figsize = (10,5))
#plot data
ax1.plot(x,x/2)
ax1.plot(x,x*2);
ax2.scatter(x,x**2);
ax3.barh(list(soft_drinks_prices.keys()),list(soft_drinks_prices.values()));
ax4.hist(student_height);
```

```python
#Option2
plt.style.use('classic')
fig,ax= plt.subplots(nrows = 2,
                            ncols = 2,
                            figsize = (10,5))
#plot data
ax[0,0].plot(x,x/2)
ax[0,0].plot(x,x*2);
ax[0,1].scatter(x,x**2);
ax[1,0].barh(list(soft_drinks_prices.keys()),list(soft_drinks_prices.values()));
ax[1,1].hist(student_height);
```
